### Cleaning Data

```python
# Menggabungkan dataset
df = pd.concat([dingling_df, dongsi_df, gucheng_df, tiantan_df, wanliu_df], ignore_index=True)
df
```

```python
# Menggabungkan kolom tahun, bulan, hari, jam menjadi satu kolom waktu
df["datetime"] = pd.to_datetime(df[["year", "month", "day", "hour"]])
df.set_index("datetime", inplace=True)
df
```

```python
# Membuang baris yang tidak memiliki data (Null) PM2.5 karena tidak bisa digunakan untuk perhitungan atau grafik.
df = df.dropna(subset=["PM2.5"])
```

**Insight:**
- Penggabungan dataset mempermudah analis
- Siapkan analisis berbasis waktu
- Menghapusan rows yang kosong pm2.5 untuk menghindari error dan hasil salah

## Exploratory Data Analysis (EDA)

### Explore Dataframe df yang menggabungkan dataset

```python
# Mengelompokkan data berdasarkan nama stasiun
# Mengambil kolom "PM2.5" dan menghitung rata-rata polusi PM2.5 untuk setiap stasiun.
# Mengurutkan hasilnya dari yang terkecil ke terbesar, agar visualisasi lebih teratur.
avg_pm25_by_station = df.groupby("station")["PM2.5"].mean().sort_values()
print("Average PM2.5 by station:\n", avg_pm25_by_station)
```

```python
# Mengelompokkan data berdasarkan kombinasi tahun dan nama stasiun.
# Menghitung rata-rata nilai PM2.5 untuk setiap stasiun di setiap tahun.
# Mengubah hasil groupby menjadi DataFrame biasa, supaya bisa dipakai di visualisas
df["year"] = df.index.year
yearly_avg = df.groupby(["year", "station"])["PM2.5"].mean().reset_index()
df
```

```python
# Menambahkan Kolom Bulan dari Index
df["month"] = df.index.month
# Mengelompokkan Bulan Menjadi Musim
df["season"] = df["month"] % 12 // 3 + 1
#Memberi Nama Musim
season_map = {1: "Winter", 2: "Spring", 3: "Summer", 4: "Autumn"}
df["season"] = df["season"].map(season_map)
# Menghitung Rata-rata PM2.5 per Musim dan Stasiun
seasonal_avg = df.groupby(["season", "station"])["PM2.5"].mean().reset_index()
df
```

**Insight:**
- Menggabungkan kolom menjadi group untuk dicari rata-ratanya
- Visualisasi eksploratif (plot)
- Analisis per waktu (tahun, musim)
- Identifikasi pola dan perbandingan

## Visualization & Explanatory Analysis

### Pertanyaan 1: Stasiun mana yang memiliki kualitas udara terburuk dan terbaik?

```python
# Plot batang Visualisasi Rata-Rata PM2.5 per Stasiun
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_pm25_by_station.index, y=avg_pm25_by_station.values)
plt.title("Rata-rata PM2.5 per Stasiun (20132017)")
plt.ylabel("Rata-rata PM2.5 (g/m)")
plt.xlabel("Station")
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Bagaimana kualitas udara dari tahun ke tahun?

```python
# Perkembangan Pulusi Tahunan PM2.5
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_avg, x="year", y="PM2.5", hue="station", marker="o")
plt.title("Rata-rata Tahunan PM2.5 by Station")
plt.ylabel("Rata-Rata PM2.5 (g/m)")
plt.xlabel("Tahun")
plt.tight_layout()
plt.show()
```

### Pertanyaan 3: Di musim apa polusi udara cenderung meningkat?


```python
# Polusi Berdasarkan Musim
plt.figure(figsize=(10, 6))
sns.barplot(data=seasonal_avg, x="season", y="PM2.5", hue="station")
plt.title("Average PM2.5 by Season and Station")
plt.ylabel("Rata-Rata PM2.5 (g/m)")
plt.xlabel("Musim")
plt.tight_layout()
plt.show()
```

**Insight:**
- Memperlihatkan perbandingan kualitas udara di kelima stasiun.
- Memberi insight apakah ada peningkatan kebijakan atau perubahan lingkungan.
- Stiap musim pengaruh polusi sama di berbagai tempat
